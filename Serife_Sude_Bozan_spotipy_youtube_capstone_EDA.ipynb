```python
# Sanatçıların frekansını hesapla
artist_counts = df['artist'].value_counts()

# Grupları tutmak için bir sözlük
grouped_counts = {'<= 5': 0, '5-10': 0, '= 10': 0}

# Koşullu kontroller
for count in artist_counts:
    if count <= 5:
        grouped_counts['<= 5'] += 1
    elif 5 < count < 10:
        grouped_counts['5-10'] += 1
    else:
        grouped_counts['= 10'] += 1

# Sonuçları bir pandas serisi olarak dönüştür
grouped_data = pd.Series(grouped_counts)

# Bar grafiği oluşturma
colors = ['#1D1C32', '#35B893', '#F55389']
plt.figure(figsize=(5,4))

ax = sns.barplot(x=grouped_data.index, y=grouped_data.values, palette=colors, width=0.4)
ax.set_facecolor('#D0D1D9')
# Sütunların üzerine frekans sayılarını ekleme
for i, value in enumerate(grouped_data.values):
    ax.text(i, value + 0.5, str(value), ha='center', va='bottom', fontsize=9)

# Başlık ve etiketler
plt.title('Artist Frequency Groups', fontsize=12)
plt.xlabel('Artist Frequency Group', fontsize=9)
plt.ylabel('Number of Artists', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
channel_counts = df['channel'].value_counts()
grouped_counts = {'<= 50': 0, '50-100': 0, '> 100': 0}
for count in channel_counts:
    if count <= 5:
        grouped_counts['<= 50'] += 1
    elif 50 < count < 100:
        grouped_counts['50-100'] += 1
    else:
        grouped_counts['> 100'] += 1
        
grouped_chan = pd.Series(grouped_counts)
```

```python
# Gruplara göre 1 atamak için bir sözlük
grouped_counts = {'< 50': 0, '50-100': 0, '> 100': 0}

# Frekansa göre gruplama ve sayma
for count in channel_counts:
    if count < 50:
        grouped_counts['< 50'] += 1  # Frekansı 50 veya daha az olanları say
    elif 50 <= count < 100:
        grouped_counts['50-100'] += 1  # Frekansı 51-99 arasında olanları say
    else:
        grouped_counts['> 100'] += 1  # Frekansı 100 ve üzeri olanları say

# Sonuçları yazdır
print("Gruplara Göre Kanal Sayıları:")
print(grouped_counts)

# Her grup için 1'lerin toplamını bul
total_count = {key: value for key, value in grouped_counts.items()}
total_count['Total'] = sum(total_count.values())

# Sonuçları yazdır
print("\nToplam Kanal Sayısı:")
print(total_count)
```

```python
# Sonuçları pandas serisine çevir
grouped_series = pd.Series(grouped_counts)

# Sonuçları yazdır
print("Gruplara Göre Kanal Sayıları:")
print(grouped_series)
color_ch_g='#05DBF2','#024873','#05F2DB'
plt.figure(figsize=(9, 3))
ax = sns.barplot(x=grouped_series.values, y=grouped_series.index, palette=color_ch_g , width=0.4)
ax.set_facecolor('#D0D1D9')  

# Barların üzerine frekans sayılarını ekleme
for i, value in enumerate(grouped_series.values):
    ax.text(value + 0.3, i, str(value), va='center', fontsize=10)

# Başlık ve etiketler
plt.title('Channel Frequency Distribution', fontsize=12)
plt.xlabel('Number of Channels', fontsize=9)
plt.ylabel('Channel Groups', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
album_counts=df['album'].value_counts()
album_counts
```

```python
# Gruplara göre 1 atamak için bir sözlük
grouped_counts_album = {'=1':0,'<= 5': 0,'<= 10': 0, '10-20': 0, '>20': 0}

# Frekansa göre gruplama ve sayma
for count in album_counts:
    if count == 1:
        grouped_counts_album['=1'] += 1  
    elif 1 < count <= 5:
        grouped_counts_album['<= 5'] += 1  
    elif 5 < count <= 10:
        grouped_counts_album['<= 10'] += 1
    elif 10 < count <= 20:
        grouped_counts_album['10-20'] += 1
    else:
        grouped_counts_album['>20'] += 1  

# Sonuçları yazdır
print("Gruplara Göre Kanal Sayıları:")
print(grouped_counts_album)

# Her grup için 1'lerin toplamını bul
total_count = {key: value for key, value in grouped_counts_album.items()}
total_count['Total'] = sum(total_count.values())

# Sonuçları yazdır
print("\nToplam Kanal Sayısı:")
print(total_count)
```

```python
# Sonuçları pandas serisine çevir
grouped_series_album = pd.Series(grouped_counts_album)

# Sonuçları yazdır
print("Gruplara Göre Kanal Sayıları:")
print(grouped_series_album)
color_al_g='#FF9561','#32C5BB','#D0EEF9'
plt.figure(figsize=(9, 4))
ax = sns.barplot(x=grouped_series_album.values, y=grouped_series_album.index, palette=color_al_g , width=0.5)
ax.set_facecolor('#D0D1D9')  

# Barların üzerine frekans sayılarını ekleme
for i, value in enumerate(grouped_series_album.values):
    ax.text(value + 0.3, i, str(value), va='center', fontsize=10)

# Başlık ve etiketler
plt.title('Album Frequency Distribution', fontsize=12)
plt.xlabel('Number of Album', fontsize=9)
plt.ylabel('Album Groups', fontsize=9)

plt.tight_layout()
plt.show()
```
